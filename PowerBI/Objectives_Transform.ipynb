```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/Oldmage112/TI4-Project/refs/heads/main/PowerBI/Derived%20CSVs/Objectives.csv'
df = pd.read_csv(url, low_memory=False, index_col= 0)
df.columns
```

```python
df
```

```python
for column in df.columns:
    df[column] = df[column].str.replace('*', '', regex=False)
```

```python
dummies = pd.get_dummies(df, columns=['PO1.1',
       'PO1.2', 'PO1.3', 'PO1.4', 'PO1.5', 'PO1.6', 'PO1.7', 'PO1.8', 'PO1.9',
       'PO1.10', 'PO1.11', 'PO1.12', 'PO1.13', 'PO1.14', 'PO1.15', 'PO1.16',
       'PO1.17', 'PO1.18', 'PO1.19', 'PO1.20', 'PO2.1', 'PO2.2', 'PO2.3',
       'PO2.4', 'PO2.5', 'PO2.6', 'PO2.7', 'PO2.8', 'PO2.9', 'PO2.10',
       'PO2.11', 'PO2.12', 'PO2.13', 'PO2.14', 'PO2.15', 'PO2.16', 'PO2.17',
       'PO2.18', 'PO2.19', 'PO2.20'], prefix='', prefix_sep='')
final_dummies = dummies.iloc[:,2:].T.groupby(level=0).sum().T
final = pd.merge(dummies.iloc[:,0:2], final_dummies, left_index=True, right_index=True)
final.to_csv("C:\GitHub\TI4 Project\PowerBI\Derived CSVs\Objectives_Transform.csv")
```
