```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/Oldmage112/TI4-Project/refs/heads/main/PowerBI/Derived%20CSVs/Objectives.csv'
df = pd.read_csv(url, index_col=2)
```

```python
dummies = pd.get_dummies(df, columns=['Public_Objectives_I - List.1', 'Public_Objectives_I - List.2',
       'Public_Objectives_I - List.3', 'Public_Objectives_I - List.4',
       'Public_Objectives_I - List.5', 'Public_Objectives_I - List.6',
       'Public_Objectives_I - List.7', 'Public_Objectives_II - Copy.1',
       'Public_Objectives_II - Copy.2', 'Public_Objectives_II - Copy.3',
       'Public_Objectives_II - Copy.4', 'Public_Objectives_II - Copy.5',
       'Public_Objectives_II - Copy.6', 'Public_Objectives_II - Copy.7',
       'Public_Objectives_II - Copy.8', 'Public_Objectives_II - Copy.9',
       'Public_Objectives_II - Copy.10', 'Public_Objectives_II - Copy.11'], prefix='', prefix_sep='')
final_dummies = dummies.iloc[:,2:].T.groupby(level=0).max().T
final = pd.merge(dummies.iloc[:,0:2], final_dummies, left_index=True, right_index=True)
final.to_csv("C:\GitHub\TI4 Project\PowerBI\Derived CSVs\Objectives_Transform.csv")
```
