```python
import pandas as pd
```

```python
url = 'https://raw.githubusercontent.com/Oldmage112/TI4-Project/refs/heads/main/PowerBI/Derived%20CSVs/Technologies.csv'
df = pd.read_csv(url, low_memory=False, index_col= -1)
df = df.fillna('')
```

```python
df
```

```python
dummies = pd.get_dummies(df, columns=['Techs.1', 'Techs.2', 'Techs.3', 'Techs.4', 'Techs.5',
       'Techs.6', 'Techs.7', 'Techs.8', 'Techs.9', 'Techs.10', 'Techs.11',
       'Techs.12', 'Techs.13', 'Techs.14', 'Techs.15', 'Techs.16', 'Techs.17',
       'Techs.18'], prefix='', prefix_sep='')
final_dummies = dummies.iloc[:,1:].T.groupby(level=0).sum().T
final = pd.merge(dummies.iloc[:,0:1], final_dummies, left_index=True, right_index=True)
final_dummies.to_csv("C:\GitHub\TI4-Project\PowerBI\Derived CSVs\Technologies_Transform.csv")
```
