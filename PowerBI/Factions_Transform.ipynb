```python
import pandas as pd
url = "C:\GitHub\TI4 Project\PowerBI\Derived CSVs\Factions.csv"
df = pd.read_csv(url, low_memory=False, index_col= 0)
```

```python
for column in df.columns[1:]:
    df[column] = df[column].str.replace('Il No Viroset', 'Il Na Viroset', regex=False)    
    for letter in ('The ','(',')','-','~','Council',' '):
        df[column] = df[column].str.replace(letter, '', regex=False)
```

```python
dummies = pd.get_dummies(df, columns=['Factions.1', 'Factions.2', 'Factions.3',
       'Factions.4', 'Factions.5', 'Factions.6', 'Factions.7', 'Factions.8'], prefix='', prefix_sep='')
final_dummies = dummies.iloc[:,1:].T.groupby(level=0).sum().T
final = pd.merge(dummies.iloc[:,0:1], final_dummies, left_index=True, right_index=True)
final = final.rename(columns= {'ASickeningLurch': 'A Sickening Lurch', 
                       'ArgentFlight' : 'Argent Flight',
                        'AvariceRex' : 'Avarice Rex', 
                        'BaronyofLetnev' : 'Barony of Letnev', 
                        'ClanofSaar' : 'Clan of Saar',
                        'KeleresArgent' : 'Keleres Argent', 
                        'KeleresMentak': 'Keleres Mentak', 
                        'KeleresXxcha': 'Keleres Xxcha',
                        'CrimsonRebellion' : 'Crimson Rebellion', 
                        'DeepwroughtScholarate' : 'Deepwrought Scholarate', 
                        'ElNenJanovet' : 'El Nen Janovet',
                        'EmbersofMuaat' : 'Embers of Muaat', 
                        'EmiratesofHacan': 'Emirates of Hacan', 
                        'FederationofSol' : 'Federation of Sol',
                        'GhostsofCreuss' : 'Ghosts of Creuss', 
                        'IlNaViroset' : 'Il Na Viroset',
                        'IlSaiLakoeHeraldofThorns' : 'Il Sai Lakoe, Herald of Thorns',
                        'L1Z1XMindnet': 'L1Z1X Mindnet', 
                        'LastBastion' : 'Last Bastion', 
                        'MahactGeneSorcerers' : 'Mahact Gene-Sorcerers',
                        'MentakCoalition': 'Mentak Coalition', 
                        'NaaluCollective' : 'Naalu Collective', 
                        'NaazRokhaAlliance':'Naaz-Rokha Alliance', 
                        'NekroVirus': 'Nekro Virus', 
                        'RadiantAur': 'Radiant Aur', 
                        'RalNelConsortium': 'Ral Nel Consortium', 
                        'RubyMonarch': 'Ruby Monarch',
                        'SaintofSwords' : 'Saint of Swords', 
                        "SardakkN'orr":"Sardakk N'orr", 
                        'TitansofUl':'Titans of Ul', 
                        'UniversitiesofJolNar':'Universities of Jol-Nar',
                        "Vuil'raithCabal" : "Vuil'raith Cabal", 
                        'XxchaKingdom' : 'Xxcha Kingdom', 
                        'YinBrotherhood': 'Yin Brotherhood',
                        'YssarilTribes': 'Yssaril Tribes'})
final.to_csv("C:\GitHub\TI4 Project\PowerBI\Derived CSVs\Factions_Transform.csv")
```
