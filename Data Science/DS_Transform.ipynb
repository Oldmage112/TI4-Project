```python
import pandas as pd
url = "https://raw.githubusercontent.com/Oldmage112/TI4-Project/refs/heads/main/PowerBI/Derived%20CSVs/Data%20Science%20Project.csv"
df = pd.read_csv(url, low_memory=False, index_col= -1)
```

```python
df
```

```python
for column in df.columns[4:]:
    df[column] = df[column].str.replace('Il No Viroset', 'Il Na Viroset', regex=False)    
    for letter in ('The ','(',')','-','~','Council',' '):
        df[column] = df[column].str.replace(letter, '', regex=False)
```

```python
df.columns
feeder = df.iloc[:,:-1]
winners = df.iloc[:,-1]
feeder
winners
```
